Seaborn

   Data Visulaization library



Importing the library

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Seaborn has somebuilt in datasets

```python
# total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# setting a theme for the plots
sns.set_theme()
```

```python
# visualize the data
sns.relplot(data=tips, x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
# load the iris dataset
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

Scatter plot

```python
sns.scatterplot(x='sepal_length',y='petal_length',hue='species',data=iris)
```

```python
sns.scatterplot(x='sepal_length',y='petal_width',hue='species',data=iris)
```

```python
# house price dataset
from sklearn.datasets import load_wine
wine_dataset=load_wine()
house= pd.DataFrame(wine_dataset.data,columns=wine_dataset.feature_names)
house['PRICE'] = wine_dataset.target
```

```python
print(wine_dataset)
```

```python
house.head()
```

Distribution plot

```python
sns.distplot(house['proline'])
```

Correlation:

1.   +ve correlation
2.   -ve correlation



Heat Map

```python
correlation=house.corr()
```

```python
# constructin heat map
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```
